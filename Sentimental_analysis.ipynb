```python
import pandas as pd
import re



# Function to detect special characters
def detect_special_characters(text):
    # Regex for detecting non-standard characters and symbols
    return bool(re.search(r"[^a-zA-Z0-9.,'\"?!:\s]", text))

# Function to detect irrelevant spaces
def detect_irrelevant_spaces(text):
    # Regex for detecting multiple consecutive spaces
    return bool(re.search(r"\s{2,}", text))

# Function to detect HTML tags
def detect_html_tags(text):
    # Regex for detecting HTML-like tags
    return bool(re.search(r"<[^>]+>", text))

# Applying the functions to the DataFrame
df['Special_Characters'] = df['Review'].apply(detect_special_characters)
df['Irrelevant_Spaces'] = df['Review'].apply(detect_irrelevant_spaces)
df['HTML_Tags'] = df['Review'].apply(detect_html_tags)


```

```python
import pandas as pd
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import nltk

# Download necessary NLTK resources
nltk.download('wordnet')
nltk.download('stopwords')


# Initialize Lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to preprocess text
def preprocess_text(text):
    # Lowercasing
    text = text.lower()

    # Keeping meaningful punctuation (like ! and ?) and removing other special characters
    text = re.sub(r'[^a-zA-Z0-9\s!?]', '', text)

    # Tokenization
    words = text.split()

    # Removing stopwords and Lemmatization
    lemmatized = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    return ' '.join(lemmatized)

# Apply the preprocessing function to the 'Review' column
df['Processed_Review'] = df['Review'].apply(preprocess_text)

df.head()  # Display the first few rows of the processed dataframe

df
```

```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import scipy.sparse as sp
import numpy as np 
from tensorflow.keras.optimizers import Adam
```

```python

# Without considering fraction of the dataset

# df is the DataFrame and 'Processed_Review' is the column with text data - 
texts = df['Processed_Review'].values
labels = df['Sentiment'].values

# Tokenization
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Padding
data = pad_sequences(sequences, maxlen=200)

# Encoding labels
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)

# Determine the number of unique classes
num_classes = np.max(encoded_labels) + 1  # Add 1 because encoded_labels are zero-indexed

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(data, encoded_labels, test_size=0.2, random_state=42)

# Building LSTM model
model = Sequential()
model.add(Embedding(5000, 128, input_shape=(200,)))  # Changed input_length to input_shape
model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(num_classes, activation='softmax'))  # Adjust for multiclass

# Compile the model
optimizer = Adam(learning_rate = 0.001)
model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy']) # Adjust loss for sparse representation

# Train the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=5, batch_size=128)
```

```python
#Considered fraction of the whole dataset
df_subset = df.sample(frac=0.1, random_state=42)
# df is the DataFrame and 'Review' is the column with text data - used Review Column instead of processed_review to determine how well it perform
texts = df_subset['Review'].values
labels = df_subset['Sentiment'].values

# Tokenization
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Padding
data = pad_sequences(sequences, maxlen=200)

# Encoding labels
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)

# Determine the number of unique classes
num_classes = np.max(encoded_labels) + 1  # Add 1 because encoded_labels are zero-indexed

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(data, encoded_labels, test_size=0.2, random_state=42)

# Building LSTM model
model = Sequential()
model.add(Embedding(5000, 100, input_shape=(200,)))  # Changed input_length to input_shape
model.add(LSTM(100))
model.add(Dense(num_classes, activation='softmax'))  # Adjust for multiclass

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy']) # Adjust loss for sparse representation

# Train the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=5, batch_size=64)
```
